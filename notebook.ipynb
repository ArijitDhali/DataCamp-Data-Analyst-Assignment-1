## The data
### **netflix_data.csv**
| Column | Description |
|--------|-------------|
| `show_id` | The ID of the show |

| `genre` | Show genre |

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Start coding here! Use as many cells as you like


# Filtering dataframe with type set as "Movie"
nf_sub = netflix_df[netflix_df["type"] == "Movie"]

# Filtering the 1990s movies with logical_and of numpy package 
logic_1 = np.logical_and(nf_sub["release_year"]>=1990 , nf_sub["release_year"]<2000)
movies_1990s = nf_sub[logic_1]

# Viewing the data for authenticity
print(movies_1990s.head())

# Viewing the size of rows returned
print(len(movies_1990s))
```

```python
#Fidning the most frequent movie duration saved in 'int'
duration = int(movies_1990s["duration"].median())
print("Median duration:",duration, "\nData Type:", type(duration))
```

```python
#Developing a histogram to view the distribution of the movie duration
plt.hist(movies_1990s["duration"],bins = 10)
plt.ylabel("Number of Movies")
plt.xlabel("Duration (in Minutes)")
plt.title("Distribution of Movie Durations in the 1990")
plt.show()
```

```python
# Filtering the 1990s movies with logical_and taking the conditions 1. Duration < 90 mins 2. Genre = Action 
logic_2 = np.logical_and(movies_1990s["duration"] < 90 , movies_1990s["genre"] == "Action")
movies_1990s_90 = movies_1990s[logic_2]

# Viewing the data for authenticity
print(movies_1990s_90)

# Viewing the size of rows returned
short_movie_count = len(movies_1990s_90)
print("Number of 1990s Short Action Movies:", short_movie_count)
```
